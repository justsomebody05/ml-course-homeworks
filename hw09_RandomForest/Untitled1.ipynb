```python
# Загрузка данных
df = pd.read_csv('ToyotaCorolla.csv')
df.info() # Для преобразования категориальных переменных
```

```python
to_drop = ['Id', 'Model', 'Mfg_Year', 'Mfg_Month', 'Cylinders', 'Color']
df_cleaned = df.drop(columns=to_drop)
```

```python
df_final = pd.get_dummies(df_cleaned, columns=['Fuel_Type'], drop_first=True)
```

2. Разбиение выборки и последующая нормализация

```python
X = df_final.drop('Price', axis = 1)
y = df_final['Price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print("Shape of X_train: ", X_train.shape)
print("Shape of X_test: ", X_test.shape)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

3. Написание модели и последующее обучение (Сравнение 2х моделей при глубине 5)

```python
rf_model = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
```

```python
dt_model = DecisionTreeRegressor(max_depth=5, random_state=42)
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
```

```python
print(f"random Forest r2: {r2_score(y_test, y_pred):.4f}, mae: {mean_absolute_error(y_test, y_pred):.2f}")
print(f"decision Tree r2: {r2_score(y_test, dt_pred):.4f}, mae: {mean_absolute_error(y_test, dt_pred):.2f}")
```

4. Оценка качества и сравнение моделей (Случайный лес и линейная регрессия)

Сравнение показало, что случайный лес стал лидером по точности, опередив линейную регрессию и одиночное дерево. Это объясняется способностью леса усреднять ошибки множества деревьев и находить сложные нелинейные связи. Одиночное дерево обучается быстрее всех, но проигрывает в точности из-за склонности к переобучению. Линейная регрессия показала стабильный результат, но уступила лесу в способности учитывать взаимодействие признаков. Таким образом, для достижения максимальной точности был выбран случайный лес.
